```python
import pandas as pd
factories = pd.read_csv('/Users/yamadatakuya/Desktop/dataknock_sample/6章/tbl_factory.csv')
factories
```

```python
warehouses = pd.read_csv('/Users/yamadatakuya/Desktop/dataknock_sample/6章/tbl_warehouse.csv')
warehouses
```

```python
cost = pd.read_csv('/Users/yamadatakuya/Desktop/dataknock_sample/6章/rel_cost.csv')
cost
```

```python
trans = pd.read_csv('/Users/yamadatakuya/Desktop/dataknock_sample/6章/tbl_transaction.csv')
trans.head()
```

```python
join_data = pd.merge(trans, cost, left_on=['ToFC', 'FromWH'], right_on=['FCID', 'WHID'], how='left')
join_data.head()
```

```python
join_data = pd.merge(join_data, factories, left_on='ToFC', right_on='FCID', how='left')
join_data.head()
```

```python
join_data = pd.merge(join_data, warehouses, left_on='FromWH', right_on='WHID', how='left')
join_data.head()
```

```python
join_data = join_data[['TransactionDate', 'Quantity', 'Cost', 'ToFC', 'FCName', 'FCDemand', 'FromWH', 'WHName', 'WHSupply', 'WHRegion']]
join_data.head()
```

```python
kanto = join_data.loc[join_data['WHRegion']=='関東']
kanto.head()
```

```python
tohoku = join_data.loc[join_data['WHRegion']=='東北']
tohoku.head()
```

```python
print('関東支社の総コスト: ' + str(kanto['Cost'].sum()) + '万円')
print('東北支社の総コスト: ' + str(tohoku['Cost'].sum()) + '万円')
```

```python
tmp = (kanto['Cost'].sum() / kanto['Quantity'].sum()) * 10000
print('関東支社の部品一つあたりの輸送コスト' + str(int(tmp)) + '円')
tmp = (tohoku['Cost'].sum() / tohoku['Quantity'].sum()) * 10000
print('東北支社の部品一つあたりの輸送コスト' + str(int(tmp)) + '円')
```

```python
cost_chk = pd.merge(cost, factories, on='FCID', how='left')
print('関東支社の平均輸送コスト' + str(cost_chk['Cost'].loc[cost_chk['FCRegion']=='関東'].mean()) + '万円')
print('東北支社の平均輸送コスト' + str(cost_chk['Cost'].loc[cost_chk['FCRegion']=='東北'].mean()) + '万円')
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_node('nodeA')
G.add_node('nodeB')
G.add_node('nodeC')

G.add_edge('nodeA', 'nodeB')
G.add_edge('nodeB', 'nodeC')
G.add_edge('nodeC', 'nodeA')

pos = {}

pos['nodeA']= (0, 0)
pos['nodeB']= (1, 1)
pos['nodeC']= (0, 1)

nx.draw(G, pos)

plt.show()
```

```python
G.add_node('nodeD')

G.add_edge('nodeA', 'nodeB')

pos['nodeD']=(1, 0)

nx.draw(G, pos, with_labels=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

df_w = pd.read_csv('/Users/yamadatakuya/Desktop/dataknock_sample/6章/network_weight.csv')
df_p = pd.read_csv('/Users/yamadatakuya/Desktop/dataknock_sample/6章/network_pos.csv')

size = 10
edge_weights = []
for i in range(len(df_w)):
    for j in range(len(df_w.columns)):
        edge_weights.append(df_w.iloc[i][j]*size)

G = nx.Graph()

for i in range(len(df_w.columns)):
    G.add_node(df_w.columns[i])

for i in range(len(df_w.columns)):
    for j in range(len(df_w.columns)):
        G.add_edge(df_w.columns[i], df_w.columns[j])

pos = {}
for i in range(len(df_w.columns)):
    node = df_w.columns[i]
    pos[node] = (df_p[node][0], df_p[node][1])
    
nx.draw(G, pos, with_labels=True, font_size=16, node_size=1000, node_color='k', font_color='w', width=edge_weights)

plt.show()
```
